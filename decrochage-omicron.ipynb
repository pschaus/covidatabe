```python
import json
import pandas as pd
import plotly.express as px

import numpy as np

import pandas as pd
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
from scipy import stats
import numpy as np
import math
df_prov_tot = pd.read_csv('static/csv/be-covid-provinces-all.csv')
df_prov_tot.dropna(inplace=True) 
df = df_prov_tot.groupby(['DATE']).agg({'TESTS_ALL': 'sum', 'TESTS_ALL_POS': 'sum','CASES': 'sum', 'NEW_IN': 'sum', 'TOTAL_IN_ICU': 'sum','TOTAL_IN_RESP': 'sum','TOTAL_IN_ECMO': 'sum'})

```

# Admissions vs ICU

```python
def bar_new_in_icu_tot():

    
    fig_hospi = make_subplots(specs=[[{"secondary_y": True, }]], shared_yaxes='all', shared_xaxes='all')

    totin_bar = go.Scatter(x=df.index, y=df.TOTAL_IN_ICU, name='#Tot ICU',mode='lines')
    fig_hospi.add_trace(totin_bar, secondary_y=True)
    
    newin_bar = go.Scatter(x=df.index, y=df.NEW_IN.rolling(7,center=True).mean(), name='#NEW IN avg(7)',mode='lines')
    fig_hospi.add_trace(newin_bar, secondary_y=False)


    fig_hospi.update_layout(template="plotly_white", height=500, margin=dict(l=0, r=0, t=30, b=0),
                            title="NEW_IN vs ICU")

    fig_hospi.update_layout(xaxis_title='Day',
                            yaxis_title='Number of / Day')
    
    fig_hospi.update_yaxes(title_text="NEW_IN", secondary_y=False)
    fig_hospi.update_yaxes(title_text="TOTAL_IN_ICU", secondary_y=True)
    

    fig_hospi.update_layout(
        hovermode='x unified',
        updatemenus=[
            dict(
                type="buttons",
                direction="left",
                buttons=list([
                    dict(
                        args=[{"yaxis.type": "linear"}],
                        label="LINEAR",
                        method="relayout"
                    ),
                    dict(
                        args=[{"yaxis.type": "log"}],
                        label="LOG",
                        method="relayout"
                    )
                ]),
            ),
        ])

    return fig_hospi

bar_new_in_icu_tot()
```

# Admissions vs Positive Rate

```python
def pos_rate_admission():

   
    fig_hospi = make_subplots(specs=[[{"secondary_y": True, }]], shared_yaxes='all', shared_xaxes='all')

    totin_bar = go.Scatter(x=df.index, y=df.TESTS_ALL_POS.rolling(7,center=True).mean()/df.TESTS_ALL.rolling(7,center=True).mean(), name='POSITIVE RATE',mode='lines')
    fig_hospi.add_trace(totin_bar, secondary_y=True)
    
    newin_bar = go.Scatter(x=df.index, y=df.NEW_IN.rolling(7,center=True).mean(), name='#NEW IN avg(7)',mode='lines')
    fig_hospi.add_trace(newin_bar, secondary_y=False)


    fig_hospi.update_layout(template="plotly_white", height=500, margin=dict(l=0, r=0, t=30, b=0),
                            title="NEW_IN vs POSITIVE RATE")

    fig_hospi.update_layout(xaxis_title='Day',
                            yaxis_title='Number of / Day')
    
    fig_hospi.update_yaxes(title_text="NEW_IN", secondary_y=False)
    fig_hospi.update_yaxes(title_text="POSITIVE RATE", secondary_y=True)
    

    fig_hospi.update_layout(
        hovermode='x unified',
        updatemenus=[
            dict(
                type="buttons",
                direction="left",
                buttons=list([
                    dict(
                        args=[{"yaxis.type": "linear"}],
                        label="LINEAR",
                        method="relayout"
                    ),
                    dict(
                        args=[{"yaxis.type": "log"}],
                        label="LOG",
                        method="relayout"
                    )
                ]),
            ),
        ])

    return fig_hospi

pos_rate_admission()
```
